```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sfla/Data_Hard.csv')
df1 = pd.read_csv('./fso/Data_Hard.csv')
```

```python
cols = ['Paths', 'Bin_Efficiency_fso', 'No_of_bins_fso', 'time_fso']
cols1 = ['Paths', 'Bin_Efficiency_sfla', 'No_of_bins_sfla', 'time_sfla']
```

```python
df1.columns = cols
```

```python
df.columns = cols1
```

```python
df['Paths'] = df['Paths'].apply(lambda x: x.split('/')[-1].split(".")[0])
```

```python
df1['Paths'] = df1['Paths'].apply(lambda x: x.split('/')[-1].split(".")[0])
```

```python
df
```

```python
df_combined = df.set_index('Paths').join(df1.set_index('Paths'))
```

```python
df_combined
```

```python
df_combined.to_csv("combined_hard.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.25
br1 = np.arange(len(df))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]


fig = plt.figure(figsize=(20,8))
plt.bar(br1, df_combined['No_of_bins_sfla'], width = barWidth)
plt.bar(br2, df_combined['No_of_bins_fso'], width = barWidth)
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(df))],
        df.Paths)
plt.yticks(np.arange(0, 100, 10));
```
